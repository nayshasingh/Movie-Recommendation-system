```python
# convert this to a different form
# ['action','adventure','fantasy','scifi']
```

```python
def convert(obj):
    L=[]
    for i in obj:
        L.append(i['name'])
    return L
```

```python
# now we need to convert this string of list to a list
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
def convertcast(obj):
    L=[]
    for i in range(3):
        if i<len(ast.literal_eval(obj)):
            L.append(ast.literal_eval(obj)[i]['name'])
    return L
```

```python
def extract(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

### remove the spaces between words to make it a single entity

```python
movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
final_df=movies[['movie_id','title','tags']]
```

```python
final_df.head()
```

```python
final_df['tags'].apply(lambda x:" ".join(x))
```

```python
# convert to a string
final_df['tags']=final_df['tags'].apply(lambda x:" ".join(x))
```

```python
final_df.head()
```

```python
final_df['tags'][0]
```

```python
# convert all in lowercase letters
final_df['tags']=final_df['tags'].apply(lambda x:x.lower())
```

```python
final_df['tags'][0]
```

```python
# do text vectorization
final_df.shape
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
stem(final_df['tags'][0])
```

```python
final_df['tags']=final_df['tags'].apply(stem)
```

```python
final_df['tags']
```

```python
# we will use bag of words 
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(final_df['tags'])
```

```python
vectors
```

```python
array_representation=vectors.toarray()
```

```python
array_representation
```

```python
cv.get_feature_names_out()
```

```python
# calculate the similarity between these vectors
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(array_representation)
```

```python
similarity
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=final_df[final_df['title']==movie].index[0]
    distances=similarity[movie_index]
    #sort these distances and extract the top 5 movies
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
#         print(i[0])
        print(final_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```
